```python
import re
import mistune
import os
import pandas as pd
import bibtexparser
import requests
```

```python
current_directory = os.getcwd()
print("Current Working Directory:", current_directory)
```

```python
# Specify the path to your .md file
file_path = "PublicationsList.md"

# Use a different variable name to avoid conflicts
with open(file_path, "r", encoding="utf-8") as md_file:
    content = md_file.read()

print(content)
```

```python
# Parse Markdown content
markdown = mistune.create_markdown()
parsed = markdown(content)

# Print the parsed output (can be modified as per need)
print(parsed)
```

```python
urls = re.findall(r'https:\/\/doi\.org\/[0-9.]+\/[a-zA-Z0-9.]+', content)

print(urls)
```

```python
urls_2 = re.findall(r'https:\/\/doi\.org\/[0-9.]+\/[a-zA-Z0-9.-]+(?:\/[a-zA-Z0-9.-]+)?', content)

df = pd.DataFrame(urls_2, columns=["doi"])
print(df)
```

```python
# Remove the URL part
dois = [re.findall(r'https:\/\/doi\.org\/([a-zA-Z0-9./-]+)', url) for url in urls_2]

# Flatten the list of lists
dois = [item[0] for item in dois if item]  # This removes any empty results

print(dois)
```

```python
# Base URL for DOI requests
doi_url_base = "https://doi.org/"

# Fetch and parse BibTeX for each DOI
for doi in dois:
    try:
        # Send a request with the Accept header set to "application/x-bibtex"
        response = requests.get(doi_url_base + doi, headers={"Accept": "application/x-bibtex"})
        response.raise_for_status()  # Raise an exception if there's an HTTP error

        # Retrieve the BibTeX string from the response
        bibtex_str = response.text

        # Parse BibTeX data using bibtexparser
        bibtex_data = bibtexparser.loads(bibtex_str)
        
        # Define a filename based on the DOI (replacing "/" with "_")
        filename = f"{doi.replace('/', '_')}.bib"

        # Save the BibTeX data to a file
        with open(filename, "w") as bibfile:
            bibfile.write(bibtex_str)
        
        print(f"Saved BibTeX for {doi} to {filename}")
        
    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve BibTeX for {doi}: {e}")
    except Exception as e:
        print(f"Failed to parse BibTeX for {doi}: {e}")
```

```python
# Base URL for DOI requests
doi_url_base = "https://doi.org/"

# List to store each BibTeX entry as a string
bibtex_entries = []

# Fetch and collect BibTeX for each DOI
for doi in dois:
    try:
        # Send request to get BibTeX data
        response = requests.get(doi_url_base + doi, headers={"Accept": "application/x-bibtex"})
        response.raise_for_status()  # Check for HTTP errors

        # Append BibTeX entry to the list
        bibtex_entries.append(response.text)

    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve BibTeX for {doi}: {e}")

# Save all entries to a single .bib file
with open("all_entries.bib", "w") as bibfile:
    for entry in bibtex_entries:
        bibfile.write(entry + "\n\n")  # Add a newline between entries

print("All BibTeX entries saved to 'all_entries.bib'")
```

```python
with open("all_entries.bib") as bibfile:
    bib_database = bibtexparser.load(bibfile)

num_entries = len(bib_database.entries)
print(f"The number of entries in all_entries.bib is: {num_entries}")

num_doi = len(dois)
print(f"The number of dois from the wiki is: {num_doi}")
```
